### 3.2  Categorical Variables Analysis

````python
# Categorical Variables Analysis
print(" CATEGORICAL VARIABLES ANALYSIS")
print("="*60)

categorical_cols = ['SEX', 'EDUCATION', 'MARRIAGE', 'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']

# Check unique values
for col in categorical_cols:
    print(f"\n {col} - Unique values:")
    value_counts = df[col].value_counts().sort_index()
    print(value_counts)
    
    # Check for unexpected values
    if col == 'SEX':
        expected = [1, 2]
        unexpected = [v for v in value_counts.index if v not in expected]
        if unexpected:
            print(f"   ⚠️  Unexpected values: {unexpected}")
    elif col == 'EDUCATION':
        expected = [1, 2, 3, 4]  # According to documentation
        unexpected = [v for v in value_counts.index if v not in expected]
        if unexpected:
            print(f"   ⚠️  Unexpected values: {unexpected} (will need cleaning)")
    elif col == 'MARRIAGE':
        expected = [1, 2, 3]  # According to documentation
        unexpected = [v for v in value_counts.index if v not in expected]
        if unexpected:
            print(f"   ⚠️  Unexpected values: {unexpected} (will need cleaning)")

# Visualize categorical variables
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.flatten()

for i, col in enumerate(categorical_cols):
    # Count plot
    sns.countplot(data=df, x=col, hue=target_col, ax=axes[i])
    axes[i].set_title(f'{col} Distribution by Default Status', fontweight='bold')
    axes[i].legend(['No Default', 'Default'])
    
    # Proportion plot
    crosstab = pd.crosstab(df[col], df[target_col], normalize='index')
    crosstab.plot(kind='bar', ax=axes[i+3], width=0.8)
    axes[i+3].set_title(f'{col} Default Rates', fontweight='bold')
    axes[i+3].set_ylabel('Proportion')
    axes[i+3].legend(['No Default', 'Default'])
    axes[i+3].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()

# Statistical significance testing (Chi-square test)
print(f"\n📊 STATISTICAL SIGNIFICANCE TESTING (Chi-square):")
for col in categorical_cols:
    contingency_table = pd.crosstab(df[col], df[target_col])
    chi2, p_value, dof, expected = chi2_contingency(contingency_table)
    
    print(f"\n{col}:")
    print(f"   Chi-square statistic: {chi2:.4f}")
    print(f"   p-value: {p_value:.4e}")
    print(f"   Degrees of freedom: {dof}")
    
    if p_value < 0.001:
        significance = "*** (Highly significant)"
    elif p_value < 0.01:
        significance = "** (Very significant)"
    elif p_value < 0.05:
        significance = "* (Significant)"
    else:
        significance = "(Not significant)"
    
    print(f"   Result: {significance}")
    
    # Calculate Cramér's V for effect size
    n = contingency_table.sum().sum()
    cramers_v = np.sqrt(chi2 / (n * (min(contingency_table.shape) - 1)))
    print(f"   Cramér's V (effect size): {cramers_v:.4f}")
````

````python
# Numerical Variables Distribution Analysis
print("🔢 NUMERICAL VARIABLES DISTRIBUTION")
print("="*60)

# Key numerical columns to analyze
key_numeric_cols = ['LIMIT_BAL', 'AGE', 'BILL_AMT1', 'BILL_AMT2', 'PAY_AMT1', 'PAY_AMT2']

# Distribution plots
fig, axes = plt.subplots(3, 4, figsize=(20, 15))
axes = axes.flatten()

for i, col in enumerate(key_numeric_cols):
    # Histogram
    axes[i*2].hist(df[col], bins=50, alpha=0.7, color='skyblue', edgecolor='black')
    axes[i*2].set_title(f'{col} - Distribution', fontweight='bold')
    axes[i*2].set_xlabel(col)
    axes[i*2].set_ylabel('Frequency')
    
    # Box plot by target
    df.boxplot(column=col, by=target_col, ax=axes[i*2+1])
    axes[i*2+1].set_title(f'{col} - by Default Status')
    axes[i*2+1].set_xlabel('Default Status')
    axes[i*2+1].set_ylabel(col)

plt.suptitle('📊 Numerical Variables Analysis', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()

# Statistical analysis
print(f"\n📊 STATISTICAL SUMMARY OF KEY VARIABLES:")
for col in key_numeric_cols:
    print(f"\n{col}:")
    stats_by_target = df.groupby(target_col)[col].agg(['count', 'mean', 'median', 'std', 'min', 'max'])
    print(stats_by_target.round(2))
    
    # Skewness and kurtosis
    skewness = df[col].skew()
    kurtosis = df[col].kurtosis()
    print(f"   Skewness: {skewness:.3f} {'(Right-skewed)' if skewness > 0.5 else '(Left-skewed)' if skewness < -0.5 else '(Normal)'}")
    print(f"   Kurtosis: {kurtosis:.3f} {'(Heavy-tailed)' if kurtosis > 3 else '(Light-tailed)' if kurtosis < 3 else '(Normal)'}")

# Outlier detection using IQR method
print(f"\n🚨 OUTLIER DETECTION (IQR Method):")
for col in key_numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_percentage = len(outliers) / len(df) * 100
    
    print(f"\n{col}:")
    print(f"   Outliers: {len(outliers):,} ({outlier_percentage:.2f}%)")
    print(f"   Range: [{lower_bound:.2f}, {upper_bound:.2f}]")
    if outlier_percentage > 5:
        print(f"   ⚠️  High outlier rate - consider capping or transformation")
````

## 5️⃣ Feature Engineering

Let's create meaningful features that will help our models better predict credit card defaults.

````python
# Feature Engineering
print("🔧 FEATURE ENGINEERING")
print("="*60)

# 1. Credit Utilization Features
print("1️⃣ Creating Credit Utilization Features...")
for i in range(1, 7):
    df_clean[f'utilization_{i}'] = np.where(
        df_clean['LIMIT_BAL'] > 0,
        df_clean[f'BILL_AMT{i}'] / df_clean['LIMIT_BAL'],
        0
    )

# Utilization aggregations
utilization_cols = [f'utilization_{i}' for i in range(1, 7)]
df_clean['utilization_avg'] = df_clean[utilization_cols].mean(axis=1)
df_clean['utilization_max'] = df_clean[utilization_cols].max(axis=1)
df_clean['utilization_std'] = df_clean[utilization_cols].std(axis=1).fillna(0)
df_clean['utilization_trend'] = df_clean['utilization_1'] - df_clean['utilization_6']  # Recent vs old

print(f"   ✅ Created {len(utilization_cols) + 4} utilization features")

# 2. Payment Behavior Features
print("2️⃣ Creating Payment Behavior Features...")
pay_status_cols = ['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']

# Delinquency analysis
df_clean['delinquency_count'] = (df_clean[pay_status_cols] > 0).sum(axis=1)
df_clean['delinquency_max'] = df_clean[pay_status_cols].max(axis=1)
df_clean['delinquency_avg'] = df_clean[pay_status_cols].mean(axis=1)

# Recent vs historical payment behavior
df_clean['recent_delinquency'] = (df_clean[['PAY_0', 'PAY_2']] > 0).sum(axis=1)  # Last 2 months
df_clean['historical_delinquency'] = (df_clean[['PAY_4', 'PAY_5', 'PAY_6']] > 0).sum(axis=1)  # 4-6 months ago

print(f"   ✅ Created 5 payment behavior features")

# 3. Payment Ratio Features
print("3️⃣ Creating Payment Ratio Features...")
for i in range(1, 7):
    bill_col = f'BILL_AMT{i}'
    pay_col = f'PAY_AMT{i}'
    df_clean[f'payment_ratio_{i}'] = np.where(
        df_clean[bill_col] > 0,
        df_clean[pay_col] / df_clean[bill_col],
        0
    )

# Payment ratio aggregations
payment_ratio_cols = [f'payment_ratio_{i}' for i in range(1, 7)]
df_clean['payment_ratio_avg'] = df_clean[payment_ratio_cols].mean(axis=1)
df_clean['payment_ratio_min'] = df_clean[payment_ratio_cols].min(axis=1)

print(f"   ✅ Created {len(payment_ratio_cols) + 2} payment ratio features")

# 4. Bill Amount Features
print("4️⃣ Creating Bill Amount Features...")
bill_cols = [f'BILL_AMT{i}' for i in range(1, 7)]
df_clean['bill_amt_avg'] = df_clean[bill_cols].mean(axis=1)
df_clean['bill_amt_std'] = df_clean[bill_cols].std(axis=1).fillna(0)
df_clean['bill_amt_trend'] = df_clean['BILL_AMT1'] - df_clean['BILL_AMT6']  # Recent vs old

print(f"   ✅ Created 3 bill amount features")

# 5. Payment Amount Features  
print("5️⃣ Creating Payment Amount Features...")
pay_amt_cols = [f'PAY_AMT{i}' for i in range(1, 7)]
df_clean['pay_amt_avg'] = df_clean[pay_amt_cols].mean(axis=1)
df_clean['pay_amt_std'] = df_clean[pay_amt_cols].std(axis=1).fillna(0)
df_clean['pay_amt_total'] = df_clean[pay_amt_cols].sum(axis=1)

print(f"   ✅ Created 3 payment amount features")

# 6. Age-based features
print("6️⃣ Creating Age-based Features...")
df_clean['age_group'] = pd.cut(df_clean['AGE'], 
                              bins=[0, 30, 40, 50, 60, 100], 
                              labels=['Young', 'Adult', 'Middle', 'Senior', 'Elder'])
# Convert to numeric for modeling
df_clean['age_group_encoded'] = df_clean['age_group'].cat.codes

print(f"   ✅ Created age group features")

# 7. Risk Score (Composite feature)
print("7️⃣ Creating Risk Score...")
df_clean['risk_score'] = (
    df_clean['utilization_avg'] * 0.3 +
    df_clean['delinquency_avg'] * 0.4 +
    (1 - df_clean['payment_ratio_avg'].clip(0, 1)) * 0.3
)

print(f"   ✅ Created composite risk score")

# Summary of new features
new_features = [col for col in df_clean.columns if col not in df.columns]
print(f"\n📊 FEATURE ENGINEERING SUMMARY:")
print(f"   Original features: {len(df.columns)}")
print(f"   New features created: {len(new_features)}")
print(f"   Total features: {len(df_clean.columns)}")

print(f"\n🔧 New Features Created:")
for i, feature in enumerate(new_features, 1):
    print(f"   {i:2d}. {feature}")

print(f"\n✅ Feature engineering completed successfully!")
````

## 6️⃣ Data Splitting & Preparation for Modeling

Let's prepare our data for machine learning by splitting it and scaling the features appropriately.

````python
# Data Splitting and Preparation
print("🔄 DATA SPLITTING & PREPARATION")
print("="*60)

# Define feature columns (excluding target and non-feature columns)
exclude_cols = [target_col, 'ID', 'age_group']  # Remove ID if present, categorical age_group
feature_cols = [col for col in df_clean.columns if col not in exclude_cols]

# Prepare features and target
X = df_clean[feature_cols].copy()
y = df_clean[target_col].copy()

print(f"📊 Dataset prepared for modeling:")
print(f"   Features: {X.shape[1]} columns")
print(f"   Samples: {X.shape[0]:,} rows")
print(f"   Target distribution: {y.value_counts().to_dict()}")

# Split the data: 70% train, 15% validation, 15% test
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.15, random_state=42, stratify=y
)

X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.176, random_state=42, stratify=y_temp  # 0.176 ≈ 0.15/0.85
)

print(f"\n🔀 Data split completed:")
print(f"   Training set:   {X_train.shape[0]:,} samples ({X_train.shape[0]/len(X):.1%})")
print(f"   Validation set: {X_val.shape[0]:,} samples ({X_val.shape[0]/len(X):.1%})")
print(f"   Test set:       {X_test.shape[0]:,} samples ({X_test.shape[0]/len(X):.1%})")

# Check class distribution in each split
for split_name, split_y in [('Training', y_train), ('Validation', y_val), ('Test', y_test)]:
    class_dist = split_y.value_counts(normalize=True)
    print(f"   {split_name} target distribution: {class_dist[0]:.3f} / {class_dist[1]:.3f}")

# Feature scaling for algorithms that need it
print(f"\n⚖️ Feature Scaling:")

# Identify numeric features for scaling
numeric_features = X_train.select_dtypes(include=[np.number]).columns.tolist()

# Create scalers
standard_scaler = StandardScaler()
robust_scaler = RobustScaler()

# Fit scalers on training data only
X_train_scaled = X_train.copy()
X_val_scaled = X_val.copy()
X_test_scaled = X_test.copy()

X_train_scaled[numeric_features] = standard_scaler.fit_transform(X_train[numeric_features])
X_val_scaled[numeric_features] = standard_scaler.transform(X_val[numeric_features])
X_test_scaled[numeric_features] = standard_scaler.transform(X_test[numeric_features])

print(f"   ✅ Standard scaling applied to {len(numeric_features)} numeric features")
print(f"   ✅ Scaling fitted on training data only (prevents data leakage)")

# Display scaling impact
scaling_comparison = pd.DataFrame({
    'Original_Mean': X_train[numeric_features].mean(),
    'Original_Std': X_train[numeric_features].std(),
    'Scaled_Mean': X_train_scaled[numeric_features].mean(),
    'Scaled_Std': X_train_scaled[numeric_features].std()
}).round(3)

print(f"\n📊 Scaling Impact (sample features):")
display(scaling_comparison.head(10))

print(f"\n✅ Data preparation completed successfully!")
````

## 7️⃣ Model Selection & Training

Now let's train multiple machine learning algorithms and compare their performance to find the best model for our credit card default prediction task.

````python
# Model Selection and Training
print("🤖 MODEL SELECTION & TRAINING")
print("="*60)

# Define models to compare (optimized for faster training)
models = {
    "Logistic Regression": LogisticRegression(random_state=42, class_weight='balanced', max_iter=500),
    "Random Forest": RandomForestClassifier(random_state=42, class_weight='balanced', n_estimators=50, n_jobs=-1),
    "XGBoost": XGBClassifier(random_state=42, eval_metric='logloss', scale_pos_weight=3.5, n_estimators=50, n_jobs=-1),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42, n_estimators=50),
    "CatBoost": CatBoostClassifier(random_state=42, class_weights=[0.22, 0.78], verbose=False, iterations=50),
    "AdaBoost": AdaBoostClassifier(random_state=42, n_estimators=50),
    "Extra Trees": ExtraTreesClassifier(random_state=42, class_weight='balanced', n_estimators=50, n_jobs=-1),
    # "SVM": SVC(random_state=42, class_weight='balanced', probability=True),  # Commented out - too slow for large datasets
}

print(f"📝 Models to train: {len(models)}")
print(f"⚡ Optimized for speed - using fewer estimators and parallel processing")

# Function to evaluate models
def evaluate_model(model, X_train, X_val, y_train, y_val, model_name):
    """Comprehensive model evaluation"""
    # Train the model
    model.fit(X_train, y_train)
    
    # Predictions
    y_pred_train = model.predict(X_train)
    y_pred_val = model.predict(X_val)
    
    # Probabilities for ROC AUC
    if hasattr(model, "predict_proba"):
        y_prob_train = model.predict_proba(X_train)[:, 1]
        y_prob_val = model.predict_proba(X_val)[:, 1]
    else:
        y_prob_train = model.decision_function(X_train)
        y_prob_val = model.decision_function(X_val)
    
    # Calculate metrics
    metrics = {
        'Model': model_name,
        'Train_Accuracy': accuracy_score(y_train, y_pred_train),
        'Val_Accuracy': accuracy_score(y_val, y_pred_val),
        'Train_Precision': precision_score(y_train, y_pred_train),
        'Val_Precision': precision_score(y_val, y_pred_val),
        'Train_Recall': recall_score(y_train, y_pred_train),
        'Val_Recall': recall_score(y_val, y_pred_val),
        'Train_F1': f1_score(y_train, y_pred_train),
        'Val_F1': f1_score(y_val, y_pred_val),
        'Train_ROC_AUC': roc_auc_score(y_train, y_prob_train),
        'Val_ROC_AUC': roc_auc_score(y_val, y_prob_val)
    }
    
    return metrics, model

# Train and evaluate all models
print("🏃‍♂️ Training models (estimated time: 2-5 minutes)...")
print("📊 Progress will be shown for each model...")
results = []
trained_models = {}

import time
total_start_time = time.time()

for i, (model_name, model) in enumerate(models.items(), 1):
    start_time = time.time()
    print(f"\n📈 [{i}/{len(models)}] Training {model_name}...", end=" ", flush=True)
    
    # Use scaled data for linear models, original for tree-based models
    if model_name in ['Logistic Regression', 'SVM']:
        X_train_use = X_train_scaled
        X_val_use = X_val_scaled
    else:
        X_train_use = X_train
        X_val_use = X_val
    
    try:
        # Add timeout for very slow models (10 minutes max per model)
        metrics, trained_model = evaluate_model(model, X_train_use, X_val_use, y_train, y_val, model_name)
        results.append(metrics)
        trained_models[model_name] = trained_model
        
        elapsed_time = time.time() - start_time
        print(f"✅ ({elapsed_time:.1f}s) - ROC-AUC: {metrics['Val_ROC_AUC']:.4f}")
        
    except Exception as e:
        elapsed_time = time.time() - start_time
        print(f"❌ Error after {elapsed_time:.1f}s: {str(e)[:30]}...")
        continue

total_elapsed_time = time.time() - total_start_time
print(f"\n⏱️  Total training time: {total_elapsed_time:.1f} seconds ({total_elapsed_time/60:.1f} minutes)")
print(f"🎯 Successfully trained: {len(results)}/{len(models)} models")

# Create results dataframe
results_df = pd.DataFrame(results)
results_df = results_df.round(4)

print(f"\n📊 MODEL COMPARISON RESULTS:")
print("="*100)
display(results_df)

# Identify best models
print(f"\n🏆 TOP PERFORMING MODELS:")
top_models = results_df.nlargest(3, 'Val_ROC_AUC')
for i, (_, row) in enumerate(top_models.iterrows(), 1):
    overfit_score = row['Train_ROC_AUC'] - row['Val_ROC_AUC']
    print(f"{i}. {row['Model']:20s} - Val ROC-AUC: {row['Val_ROC_AUC']:.4f} (Overfit: {overfit_score:.4f})")

best_model_name = top_models.iloc[0]['Model']
best_model = trained_models[best_model_name]
print(f"\n🎯 Selected Best Model: {best_model_name}")

print(f"\n✅ Model training completed successfully!")
````

## 9️⃣ Hyperparameter Tuning

Let's optimize our best model's hyperparameters to achieve even better performance.

````python
# Hyperparameter Tuning
print("🔧 HYPERPARAMETER TUNING")
print("="*60)

# Define hyperparameter grids for top models
param_grids = {
    "Random Forest": {
        'n_estimators': [100, 200, 300],
        'max_depth': [10, 20, None],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'class_weight': ['balanced', 'balanced_subsample']
    },
    "XGBoost": {
        'n_estimators': [100, 200, 300],
        'max_depth': [3, 6, 9],
        'learning_rate': [0.01, 0.1, 0.2],
        'subsample': [0.8, 0.9, 1.0],
        'scale_pos_weight': [3, 4, 5]
    },
    "Logistic Regression": {
        'C': [0.001, 0.01, 0.1, 1, 10, 100],
        'penalty': ['l1', 'l2', 'elasticnet'],
        'solver': ['liblinear', 'saga'],
        'class_weight': ['balanced', None]
    },
    "Gradient Boosting": {
        'n_estimators': [100, 200],
        'learning_rate': [0.01, 0.1, 0.2],
        'max_depth': [3, 5, 7],
        'subsample': [0.8, 0.9, 1.0]
    }
}

# Select top 2 models for tuning (to save time)
top_2_models = results_df.nlargest(2, 'Val_ROC_AUC')['Model'].tolist()
print(f"🎯 Tuning top 2 models: {top_2_models}")

tuned_results = {}

for model_name in top_2_models:
    if model_name not in param_grids:
        print(f"   ⚠️ Skipping {model_name} - no param grid defined")
        continue
        
    print(f"\n🔍 Tuning {model_name}...")
    
    # Get the base model
    if model_name == "Random Forest":
        base_model = RandomForestClassifier(random_state=42)
    elif model_name == "XGBoost":
        base_model = XGBClassifier(random_state=42, eval_metric='logloss')
    elif model_name == "Logistic Regression":
        base_model = LogisticRegression(random_state=42, max_iter=1000)
    elif model_name == "Gradient Boosting":
        base_model = GradientBoostingClassifier(random_state=42)
    
    # Select appropriate data
    if model_name in ['Logistic Regression', 'SVM']:
        X_train_tune = X_train_scaled
        X_val_tune = X_val_scaled
    else:
        X_train_tune = X_train
        X_val_tune = X_val
    
    # Combine training and validation for cross-validation
    X_tune = pd.concat([X_train_tune, X_val_tune])
    y_tune = pd.concat([y_train, y_val])
    
    # Use RandomizedSearchCV for faster tuning
    random_search = RandomizedSearchCV(
        estimator=base_model,
        param_distributions=param_grids[model_name],
        n_iter=10,  # Reduced from 20 for faster execution
        cv=3,       # 3-fold cross-validation
        scoring='roc_auc',
        random_state=42,
        n_jobs=-1,  # Use all available processors
        verbose=1   # Show progress
    )
    
    print(f"   🔄 Running {10} parameter combinations with 3-fold CV...")
    
    # Fit the random search
    random_search.fit(X_tune, y_tune)
    
    # Store results
    tuned_results[model_name] = {
        'best_model': random_search.best_estimator_,
        'best_score': random_search.best_score_,
        'best_params': random_search.best_params_
    }
    
    print(f"   ✅ Best CV ROC-AUC: {random_search.best_score_:.4f}")
    print(f"   🎛️ Best parameters: {random_search.best_params_}")

# Compare tuned vs original models
print(f"\n📊 TUNING RESULTS COMPARISON:")
print("="*80)

comparison_results = []
for model_name in top_2_models:
    if model_name in tuned_results:
        # Original model performance
        original_score = results_df[results_df['Model'] == model_name]['Val_ROC_AUC'].iloc[0]
        
        # Tuned model performance  
        tuned_score = tuned_results[model_name]['best_score']
        
        improvement = tuned_score - original_score
        
        comparison_results.append({
            'Model': model_name,
            'Original_ROC_AUC': original_score,
            'Tuned_ROC_AUC': tuned_score,
            'Improvement': improvement,
            'Improvement_%': (improvement / original_score) * 100
        })

comparison_df = pd.DataFrame(comparison_results)
display(comparison_df.round(4))

# Select the best tuned model
if comparison_results:
    best_tuned = max(comparison_results, key=lambda x: x['Tuned_ROC_AUC'])
    best_tuned_name = best_tuned['Model']
    final_model = tuned_results[best_tuned_name]['best_model']
    
    print(f"\n🏆 FINAL SELECTED MODEL:")
    print(f"   Model: {best_tuned_name}")
    print(f"   Tuned ROC-AUC: {best_tuned['Tuned_ROC_AUC']:.4f}")
    print(f"   Improvement: {best_tuned['Improvement']:.4f} ({best_tuned['Improvement_%']:+.2f}%)")
    print(f"   Best Parameters: {tuned_results[best_tuned_name]['best_params']}")
else:
    final_model = best_model
    best_tuned_name = best_model_name
    print(f"\n🎯 Using original best model: {best_model_name}")

print(f"\n✅ Hyperparameter tuning completed!")
````

Let's evaluate our final optimized model on the unseen test set to get an unbiased performance estimate.

````python
# Final Model Evaluation on Test Set
print("🏁 FINAL MODEL EVALUATION")
print("="*60)

# Use appropriate test data
if best_tuned_name in ['Logistic Regression', 'SVM']:
    X_test_final = X_test_scaled
else:
    X_test_final = X_test

# Generate final predictions
y_pred_test = final_model.predict(X_test_final)
y_prob_test = final_model.predict_proba(X_test_final)[:, 1]

# Calculate comprehensive metrics
test_metrics = {
    'Accuracy': accuracy_score(y_test, y_pred_test),
    'Precision': precision_score(y_test, y_pred_test),
    'Recall': recall_score(y_test, y_pred_test),
    'F1-Score': f1_score(y_test, y_pred_test),
    'ROC-AUC': roc_auc_score(y_test, y_prob_test),
    'Average Precision': average_precision_score(y_test, y_prob_test)
}

print(f"🎯 FINAL MODEL: {best_tuned_name}")
print(f"📊 TEST SET PERFORMANCE:")
print("-" * 40)
for metric, value in test_metrics.items():
    print(f"   {metric:18s}: {value:.4f}")

# Final confusion matrix
cm_test = confusion_matrix(y_test, y_pred_test)
tn, fp, fn, tp = cm_test.ravel()

print(f"\n📋 FINAL CONFUSION MATRIX:")
print(f"                 Predicted")
print(f"                No     Yes    Total")
print(f"Actual    No   {tn:5d}   {fp:4d}   {tn+fp:5d}")
print(f"          Yes  {fn:5d}   {tp:4d}   {fn+tp:5d}")
print(f"          Total {tn+fn:4d}   {fp+tp:4d}   {len(y_test):5d}")

# Business Impact Analysis
print(f"\n💼 BUSINESS IMPACT ANALYSIS:")
total_customers = len(y_test)
actual_defaults = sum(y_test)
predicted_defaults = sum(y_pred_test)

# Assuming average loss per default is $5000 and investigation cost is $100
avg_loss_per_default = 5000
investigation_cost = 100

# Current losses (without model)
current_total_loss = actual_defaults * avg_loss_per_default

# With model
prevented_defaults = tp
missed_defaults = fn
false_alarms = fp

savings_from_prevention = prevented_defaults * avg_loss_per_default
investigation_costs = predicted_defaults * investigation_cost
net_savings = savings_from_prevention - investigation_costs
roi_percentage = (net_savings / current_total_loss) * 100

print(f"   Total customers in test:     {total_customers:,}")
print(f"   Actual defaults:             {actual_defaults:,}")
print(f"   Correctly identified:        {tp:,} ({tp/actual_defaults:.1%})")
print(f"   Missed defaults:             {fn:,} ({fn/actual_defaults:.1%})")
print(f"   False alarms:                {fp:,}")
print(f"")
print(f"   💰 Current annual loss:      ${current_total_loss:,}")
print(f"   💰 Prevented losses:         ${savings_from_prevention:,}")
print(f"   💰 Investigation costs:      ${investigation_costs:,}")
print(f"   💰 Net annual savings:       ${net_savings:,}")
print(f"   📈 ROI:                      {roi_percentage:.1f}%")

# Model interpretation for business
if hasattr(final_model, 'feature_importances_'):
    top_features = pd.DataFrame({
        'Feature': X_train.columns,
        'Importance': final_model.feature_importances_
    }).sort_values('Importance', ascending=False).head(10)
    
    print(f"\n🔍 TOP 10 MOST IMPORTANT FEATURES:")
    for i, (_, row) in enumerate(top_features.iterrows(), 1):
        print(f"   {i:2d}. {row['Feature']:25s}: {row['Importance']:.4f}")

print(f"\n✅ Final evaluation completed successfully!")
````

```python
# 11. Fix / recode known invalid categories:
# EDUCATION valid values per UCI: 1 = graduate school; 2 = university; 3 = high school; 4 = others
# But dataset has 0, 5, 6 as well — map to 4 (others / unknown)
df_clean = df.copy()
df_clean = df_clean.rename(columns={'PAY_0': 'PAY_1'})

# Now define pay_status columns
pay_status_cols = ['PAY_1', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
# EDUCATION fix
print("Before EDUCATION value counts:\n", df_clean['EDUCATION'].value_counts().sort_index())
df_clean['EDUCATION'] = df_clean['EDUCATION'].replace({0:4, 5:4, 6:4})
print("After EDUCATION value counts:\n", df_clean['EDUCATION'].value_counts().sort_index())

# MARRIAGE fix: values should be 1,2,3 (but 0 exists -> set to 3 'others/unknown')
print("Before MARRIAGE value counts:\n", df_clean['MARRIAGE'].value_counts().sort_index())
df_clean['MARRIAGE'] = df_clean['MARRIAGE'].replace({0:3})
print("After MARRIAGE value counts:\n", df_clean['MARRIAGE'].value_counts().sort_index())
```

```python
# 15. Outlier handling suggestions & optional trimming example
# We saw LIMIT_BAL has extreme values; one simple approach is to clip at some quantile, or log-transform.
print("LIMIT_BAL stats pre-clipping:", df_clean['LIMIT_BAL'].describe())

# Example: create a clipped version and a log version for modeling later
df_clean['LIMIT_BAL_clipped'] = df_clean['LIMIT_BAL'].clip(upper=df_clean['LIMIT_BAL'].quantile(0.99))
df_clean['LIMIT_BAL_log'] = np.log1p(df_clean['LIMIT_BAL'])

print("LIMIT_BAL_clipped max:", df_clean['LIMIT_BAL_clipped'].max())
print("LIMIT_BAL_log sample:\n", df_clean['LIMIT_BAL_log'].head())
```

**Explanation:**  
- 99th percentile value used as a cap — clipping keeps most of the distribution intact while preventing a few huge values (like 1M) from dominating.  
- `LIMIT_BAL_log` compresses the numeric range and reduces skew; median on log scale ≈ 11.85 (useful for linear models).  
**Decision:** keep raw, clipped, and log versions — we will compare model performance using each representation.

Credit utilization (bill / limit) is a key measure of credit risk. Clients who consistently max out their limit are more likely to default. Instead of only utilization_1, we compute for all 6 months and also summary stats (avg, max, volatility, trend).

```python
# Code
for i in range(1,7):
    df_clean[f'utilization_{i}'] = df_clean[f'BILL_AMT{i}'] / df_clean['LIMIT_BAL']

# Aggregates
df_clean['utilization_avg'] = df_clean[[f'utilization_{i}' for i in range(1,7)]].mean(axis=1)
df_clean['utilization_max'] = df_clean[[f'utilization_{i}' for i in range(1,7)]].max(axis=1)
df_clean['utilization_std'] = df_clean[[f'utilization_{i}' for i in range(1,7)]].std(axis=1)
```

Explanation

utilization_i: % of limit used each month.

utilization_avg: mean usage across 6 months.

utilization_max: worst month (highest exposure).

utilization_std: volatility — how much their usage fluctuates.
👉 High average, high max, or volatile utilization often signal risky behavior.

We already created delinquency_count = #months with PAY_* > 0. Let’s also capture maximum delay and average delay across 6 months.

```python
# Code
df_clean['delinquency_max'] = df_clean[pay_status_cols].max(axis=1)
df_clean['delinquency_avg'] = df_clean[pay_status_cols].mean(axis=1)
```

Explanation

delinquency_max: longest delay in months → if someone has ever been 6 months late, they’re very high risk.

delinquency_avg: overall payment timeliness → higher values mean consistent lateness.
👉 These are among the most predictive features in credit scoring.

Absolute bill and payment amounts aren’t enough; what matters is how much of the bill they actually pay. This ratio shows repayment discipline.

```python
# Code
for i in range(1,7):
    df_clean[f'pay_ratio_{i}'] = df_clean[f'PAY_AMT{i}'] / (df_clean[f'BILL_AMT{i}'].replace(0, np.nan))
    
# Aggregates
df_clean['pay_ratio_avg'] = df_clean[[f'pay_ratio_{i}' for i in range(1,7)]].mean(axis=1)
df_clean['pay_ratio_max'] = df_clean[[f'pay_ratio_{i}' for i in range(1,7)]].max(axis=1)
```

Explanation

pay_ratio_i: fraction of bill repaid in month i. (NaN if bill=0; we’ll handle missing after).

pay_ratio_avg: repayment discipline across 6 months.

pay_ratio_max: best repayment month.
👉 A low average ratio → chronic underpayment → higher risk.

Total spending and repayments over 6 months give long-term financial behavior.

```python
# Code
df_clean['total_bill_6m'] = df_clean[bill_cols].sum(axis=1)
df_clean['total_pay_6m'] = df_clean[payamt_cols].sum(axis=1)
df_clean['pay_to_bill_ratio'] = df_clean['total_pay_6m'] / (df_clean['total_bill_6m'].replace(0, np.nan))
```

Explanation

total_bill_6m: overall debt exposure.

total_pay_6m: overall repayments.

pay_to_bill_ratio: ability to cover debts over 6 months.
👉 High bills but low repayments = high risk.

Bucketing converts continuous values into interpretable ranges. Banks often use buckets for risk scoring.

```python
# Code
# Age buckets
df_clean['AGE_BUCKET'] = pd.cut(df_clean['AGE'], bins=[20,30,40,50,60,70,80], labels=['20s','30s','40s','50s','60s','70s'])

# Limit buckets
df_clean['LIMIT_BUCKET'] = pd.cut(df_clean['LIMIT_BAL'], bins=[0,50000,150000,300000,500000,1000000], 
                                  labels=['very_low','low','medium','high','very_high'])
```

Explanation

AGE_BUCKET: groups clients into decades → easier to interpret age-related trends.

LIMIT_BUCKET: groups clients into credit exposure levels.
👉 Useful for reporting, and may capture nonlinear relationships.

Categorical variables (SEX, EDUCATION, MARRIAGE, buckets) need one-hot encoding for Logistic Regression.

```python
# Code
df_encoded = pd.get_dummies(df_clean, columns=['SEX','EDUCATION','MARRIAGE','AGE_BUCKET','LIMIT_BUCKET'], drop_first=True)
```

Each category becomes its own binary column.

Example: EDUCATION becomes EDUCATION_2, EDUCATION_3, EDUCATION_4.

drop_first=True prevents dummy variable trap (multicollinearity).
👉 Dataset is now fully numeric and model-ready.

```python
import pandas as pd
from sklearn.impute import SimpleImputer

# Load the engineered dataset again
file_path = 'credit_data_features.csv'
df_features = pd.read_csv(file_path)

# Define target
target_col = 'default.payment.next.month'

# Separate features and target
X = df_features.drop(columns=[target_col])
y = df_features[target_col]

# Impute missing values (from ratios where BILL_AMT=0)
imputer = SimpleImputer(strategy="constant", fill_value=0)
X_imputed = imputer.fit_transform(X)

# Convert back to DataFrame for easy handling
X = pd.DataFrame(X_imputed, columns=X.columns)

# Verify
print(X.shape, y.shape, X.isnull().sum().sum())
```

```python
# Phase 3: Modeling - Step 1 Logistic Regression (Baseline)

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
import seaborn as sns
import matplotlib.pyplot as plt

# ---------------------------
# Train-test split (stratified)
# ---------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# ---------------------------
# Logistic Regression training
# ---------------------------
log_reg = LogisticRegression(
    max_iter=1000, 
    class_weight='balanced',  # handle imbalance
    solver='lbfgs'
)
log_reg.fit(X_train, y_train)

# ---------------------------
# Predictions & probabilities
# ---------------------------
y_pred_lr = log_reg.predict(X_test)
y_proba_lr = log_reg.predict_proba(X_test)[:,1]

# ---------------------------
# Evaluation metrics
# ---------------------------
roc_auc_lr = roc_auc_score(y_test, y_proba_lr)
print("ROC-AUC Score:", round(roc_auc_lr, 3))

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred_lr, digits=3))

# ---------------------------
# Confusion Matrix Visualization
# ---------------------------
cm_lr = confusion_matrix(y_test, y_pred_lr)

plt.figure(figsize=(5,4))
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['No Default (0)', 'Default (1)'],
            yticklabels=['No Default (0)', 'Default (1)'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Logistic Regression - Confusion Matrix")
plt.show()
```

```python
# Phase 3: Modeling - Step 2 Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
import seaborn as sns
import matplotlib.pyplot as plt

# ---------------------------
# Train Random Forest
# ---------------------------
rf = RandomForestClassifier(
    n_estimators=200,        # number of trees
    max_depth=10,            # prevent overfitting by limiting tree depth
    class_weight='balanced', # handle class imbalance
    random_state=42,
    n_jobs=-1                # use all CPU cores
)
rf.fit(X_train, y_train)

# ---------------------------
# Predictions & Probabilities
# ---------------------------
y_pred_rf = rf.predict(X_test)
y_proba_rf = rf.predict_proba(X_test)[:,1]

# ---------------------------
# Evaluation Metrics
# ---------------------------
roc_auc_rf = roc_auc_score(y_test, y_proba_rf)
print("ROC-AUC Score:", round(roc_auc_rf, 3))

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred_rf, digits=3))

# ---------------------------
# Confusion Matrix Visualization
# ---------------------------
cm_rf = confusion_matrix(y_test, y_pred_rf)

plt.figure(figsize=(5,4))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Greens',
            xticklabels=['No Default (0)', 'Default (1)'],
            yticklabels=['No Default (0)', 'Default (1)'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Random Forest - Confusion Matrix")
plt.show()
```

```python
# Phase 3: Modeling - Step 3 XGBoost Classifier

from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
import seaborn as sns
import matplotlib.pyplot as plt

# ---------------------------
# Train XGBoost
# ---------------------------
xgb = XGBClassifier(
    n_estimators=300,        # number of boosting rounds
    max_depth=6,             # depth of trees
    learning_rate=0.05,      # shrinkage rate
    subsample=0.8,           # row sampling
    colsample_bytree=0.8,    # feature sampling
    scale_pos_weight=(len(y_train)-sum(y_train))/sum(y_train), # handle imbalance
    random_state=42,
    use_label_encoder=False,
    eval_metric="logloss"    # suppress warning
)
xgb.fit(X_train, y_train)

# ---------------------------
# Predictions & Probabilities
# ---------------------------
y_pred_xgb = xgb.predict(X_test)
y_proba_xgb = xgb.predict_proba(X_test)[:,1]

# ---------------------------
# Evaluation Metrics
# ---------------------------
roc_auc_xgb = roc_auc_score(y_test, y_proba_xgb)
print("ROC-AUC Score:", round(roc_auc_xgb, 3))

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred_xgb, digits=3))

# ---------------------------
# Confusion Matrix Visualization
# ---------------------------
cm_xgb = confusion_matrix(y_test, y_pred_xgb)

plt.figure(figsize=(5,4))
sns.heatmap(cm_xgb, annot=True, fmt='d', cmap='Oranges',
            xticklabels=['No Default (0)', 'Default (1)'],
            yticklabels=['No Default (0)', 'Default (1)'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("XGBoost - Confusion Matrix")
plt.show()
```
